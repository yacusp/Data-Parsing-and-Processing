# Задание 1
Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
# Импортируем библиотеки
import requests
import json
```

```python
# cсылка на github api
url = 'https://api.github.com'
# имя пользователя GitHub, чьи репозитории нас интересуют
user = 'Yacusp' 
get_list = requests.get(f'{url}/users/{user}/repos')
```

```python
# записываем полученные данные в файл result.json
with open('result.json', 'w') as f:
    json.dump(get_list.json(), f)
```

```python
# проверяем
for i in get_list.json():
    print(i['name'])
```
